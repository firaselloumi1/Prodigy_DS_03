```python
# Selecting the columns
column = df[['age', 'campaign', 'duration']]

# Calculating Q1, Q3, and IQR for each column
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1

# Calculating the lower and upper bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filtering the DataFrame to include only rows within the bounds
df[['age', 'campaign', 'duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
df.plot(kind='box', subplots=True, layout=(2, 5), figsize=(20, 10), boxprops=dict(color='#808000'))
plt.show()
```

```python
high_corr_cols = ['emp.var.rate', 'euribor3m', 'nr.employed']
```

```python
df1 = df.copy()  # Corrected the copy method
df1.columns      # Access the columns of the DataFrame
```

```python
df1.drop(high_corr_cols, inplace=True, axis=1)  # axis=1 indicates columns
print(df1.columns)  # Display the remaining columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
Ib = LabelEncoder ()
df_encoded = df1.apply(Ib.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
# Drop the 'deposit' column to create independent variable 'x'
x = df_encoded.drop('deposit', axis=1)  # independent variable

# Create dependent variable 'y'
y = df_encoded['deposit']  # dependent variable

# Print the shapes and types
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Assuming 'x' and 'y' are already defined
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

# Print the shapes of the training and testing sets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def eval_model(y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    print('Accuracy Score:', acc)
    
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:\n', cm)
    
    print('Classification Report:\n', classification_report(y_test, y_pred))

def mscore(model):
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    
    print('Training Score:', train_score)
    print('Testing Score:', test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Define a value for max_depth
s = 5  # You can change this to any integer value you want

# Create an instance of DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=s, min_samples_split=10)

# Fit the model on the training data
dt.fit(x_train, y_train)

```

```python
mscore(dt)
```

```python
# Make predictions using the trained Decision Tree model
ypred_dt = dt.predict(x_test)

# Print the predictions
print(ypred_dt)
```

```python
eval_model (y_test,ypred_dt)
```

```python
from sklearn.tree import plot_tree
```

```python
# Assuming 'cn' is the list of classes and 'fn' contains the feature names
cn = ['no', 'yes']
fn = x_train.columns

# Print the feature names
print(fn)

# If you want to print the class names, use 'cn' instead of 'en'
print(cn)
```

```python
# Set the figure size
plt.figure(figsize=(30, 10))

# Plot the decision tree
plot_tree(dt, class_names=cn, filled=True)

# Display the plot
plt.show()
```
